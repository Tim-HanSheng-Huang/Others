```python
from facebook_scraper import get_posts
import pandas as pd
import csv 
import time
```

```python
def collectposts(ID):
    posts = []
    count = 0
    start_time = time.time()
    for onepost in get_posts(ID, pages=500, extra_info=True):
        count = count + 1

        time1 = str(onepost['time'])
        if time1.startswith("2019-09"):
                break
#         if count %10 == 0:
#             time.sleep(2)

#             print("--- %s seconds ---" % (time.time() - start_time))
        print(onepost['time'])
        posts.append(onepost)
    return posts
```

```python
def todict(posts):
    outputlist=[]
    
    for onepost in posts:
        
        try:
            text = onepost['text'].replace('\n',"")
        except :
            text = onepost['text']

        try:
            shared_Text = onepost['shared_text'].replace('\n',"")       
        except :
            shared_Text = onepost['shared_text']


        #image and video
        
        image = 0
        if onepost['image'] != None:
            image = 1

        video = 0
        videos_URL = onepost['video']
        if videos_URL != None:
            video = 1
        video_id = onepost['video_id']

        #reactions
        try:
            like=onepost['reactions']['like']
        except KeyError as e:
            like=0

        try:
            support=onepost['reactions']['support']
        except KeyError as e:
            support=0

        try:
            love=onepost['reactions']['love']
        except KeyError as e:
            love=0

        try:
            wow=onepost['reactions']['wow']
        except KeyError as e:
            wow=0

        try:
            haha=onepost['reactions']['haha']
        except KeyError as e:
            haha=0
        
        onerow=[{'post_id':str(onepost['post_id']),
                 'post_url':str(onepost['post_url']),

                 'time':str(onepost['time']),
                 'text':str(text),
                 'shared_text':str(shared_Text),

                 'comments':str(onepost['comments']),
                 'shares':str(onepost['shares']),

                 'like':str(like),
                 'support':str(support),
                 'love':str(love),
                 'wow':str(wow),
                 'haha':str(haha),
                 'link':str(onepost['link']),

                 'image':str(image),
                 'image_URL':str(onepost['image']),
                 'video':str(video),
                 'video_URL':str(videos_URL),
                 'video_ID':str(video_id)}]
            
        print(onepost['time'])
        outputlist.append(onerow)
        
    return outputlist
```

```python
def saveCSV(outputlist,filename):
    
    filename = filename+'.csv'
    file = open(filename, 'w', newline ='',encoding="utf_8_sig") 

    with file: 
        # identifying header   
        header = ['post_id','post_url','time','text','shared_text',\
                  '貼文分類',\
                  'comments','shares','like','support','love','wow','haha',\
                  'link','image','image_URL','video','video_URL','video_ID'] 
        writer = csv.DictWriter(file, fieldnames = header) 
        writer.writeheader() 
        for i in range(len(outputlist)):
            # writing data row-wise into the csv file 
            writer.writerows(outputlist[i]) 
```

```python
data = pd.read_excel('candidates.xlsx', header =0) 
df = pd.DataFrame(data)
df
```

```python
for i in range(270,271):
    Section = str(df.iloc[i][0])
    NO = str(df.iloc[i][3])
    Name = str(df.iloc[i][1])
    ID = str(df.iloc[i][2])
    
    FileName = Section+'_'+NO+'_'+Name
    
    print(i,FileName)
    try:
        posts = collectposts(ID)
    except:
        print("HTTPError:",FileName)
        continue
        
    outputlist = todict(posts)
    saveCSV(outputlist,FileName)
```
