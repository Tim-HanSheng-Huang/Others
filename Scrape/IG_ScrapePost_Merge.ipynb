```python
# browser = webdriver.Chrome()  
url='https://www.instagram.com/explore/tags/台北景點/'
browser.get(url) # 前往該網址
post_list = []
```

```python
# 往下滑並取得新的貼文連結
# n_scroll = 20
# for i in range(n_scroll):
while len(post_list)<1000:
#     print(i,end=' ')
    scroll = 'window.scrollTo(0, document.body.scrollHeight);'
    browser.execute_script(scroll)
    html = browser.page_source
    soup = Soup(html, 'html.parser')

    # 尋找所有的貼文連結
    for elem in soup.select('article div div div div a'):
        # 如果新獲得的貼文連結不在列表裡，則加入
        if elem['href'] not in post_list:
            post_list.append(elem['href'])
    print('post number',str(len(post_list)))
    time.sleep(2) # 等待網頁加載

# 總共加載的貼文連結數
print("總共取得 " + str(len(post_list)) + " 篇貼文連結")
```

```python
# MyFile=open('台北景點.txt','w')

# for element in post_list:
#      MyFile.write(element)
#      MyFile.write('\n')
# MyFile.close()

post_list=[]
f = open('台北景點.txt')
for line in f.readlines():
    line.replace("\n","")
    post_list.append(line)
f.close

print(post_list)
len(post_list)
```

```python
import time
# 抓每篇資料
post_instascrape_list=[]
wait=40

for j in range(wait):
    # url個數/wait=一次跑的數量
    i=j*int(len(post_list)/wait)
    # 一次最多只能跑70篇
    for i in range(i,int(i+len(post_list)/wait)):
        print(i)
        shortcode=post_list[i]
        shortcode=str(shortcode)
#         shortcode=shortcode[2:17]
        URL="https://www.instagram.com"
        URL=URL+str(shortcode)
        time.sleep(i%3+2)
        try:
            # post物件
            post_instascrape=Post(URL)
            post_instascrape.scrape()
        except ValueError as e:
            print('ValueError:',e)
            continue
        post_instascrape_list.append(post_instascrape)
    time.sleep(8)
```

```python
outputlist=[]

for i in range(len(post_instascrape_list)):
    onepost=post_instascrape_list[i].to_dict()
    print(onepost['shortcode'])
    
#     raw_content_list=onepost['caption'].split()
#     hashtaglist=[]
#     contentlist=[]
#     for j in range(len(raw_content_list)):
#         item=raw_content_list[j]
#         if item[0]=='#':
#             hashtaglist.append(item)
#         else:
#             contentlist.append(item)

#     after_content=" ".join(contentlist)
#     after_hashtag=" ".join(hashtaglist)
#     after_content=demoji.replace(after_content,"")
#     after_hashtag=demoji.replace(after_hashtag,"")
    content=demoji.replace(onepost['caption'],"")
    
    onerow=[{'shortcode':str(onepost['shortcode']),
            'likes':str(onepost['likes']),
              'comments':str(onepost['comments']),
              'upload_date':str(onepost['upload_date']),
#               'content':str(after_content),
#               'hashtag':str(after_hashtag),
            'content':str(content)}]
    outputlist.append(onerow)
```

```python
# opening the csv file in 'w' mode 
file = open('台北景點.csv', 'w', newline ='',encoding="big5") 

with file: 
    # identifying header   
    header = ['shortcode','likes','comments','upload_date','content'] 
    writer = csv.DictWriter(file, fieldnames = header) 
    writer.writeheader() 
    for i in range(len(outputlist)):
        # writing data row-wise into the csv file 
        writer.writerows(outputlist[i]) 
```
